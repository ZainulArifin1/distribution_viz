```python
# import packages
import statsmodels.api as sm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import data

```python
#iris dataset
iris = sm.datasets.get_rdataset('iris').data

#adults dataset
adults = pd.read_csv("adults.csv")
```

# HISTOGRAM

```python
import warnings
warnings.filterwarnings('ignore')

iris_species = pd.Series(iris["Species"].unique()).tolist()

plt.figure(num=None, figsize=(20, 12), dpi=80, facecolor='w', edgecolor='k')

for x in iris_species:
    # Subset to the airline
    subset = iris[iris['Species'] == x]
    
    # Draw the density plot
    sns.distplot(subset['Petal.Length'], hist = True, kde = False,
                 kde_kws = {'linewidth': 1, 'shade' : True},
                 bins=10,
                 label = x)

    
plt.axvline(iris[iris["Species"] == "setosa"]["Petal.Length"].median(), color='blue', linestyle='solid', linewidth=2, alpha = 0.5)
plt.axvline(iris[iris["Species"] == "versicolor"]["Petal.Length"].median(), color='orange', linestyle='solid', linewidth=2, alpha = 0.5)
plt.axvline(iris[iris["Species"] == "virginica"]["Petal.Length"].median(), color='green', linestyle='solid', linewidth=2, alpha = 0.5)
plt.legend(prop={'size': 16}, title = 'IRIS PETAL LENGTH')
plt.title('Histogram of 3 Iris Species Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Freq')
```

# DENSITY PLOT

```python
iris_species = pd.Series(iris["Species"].unique()).tolist()

plt.figure(num=None, figsize=(20, 12), dpi=80, facecolor='w', edgecolor='k')

for x in iris_species:
    # Subset to the airline
    subset = iris[iris['Species'] == x]
    
    # Draw the density plot
    sns.distplot(subset['Petal.Length'], hist = False, kde = True,
                 kde_kws = {'linewidth': 1, 'shade' : True},
                 bins=10,
                 label = x)

    
plt.axvline(iris[iris["Species"] == "setosa"]["Petal.Length"].median(), color='blue', linestyle='solid', linewidth=2, alpha = 0.5)
plt.axvline(iris[iris["Species"] == "versicolor"]["Petal.Length"].median(), color='orange', linestyle='solid', linewidth=2, alpha = 0.5)
plt.axvline(iris[iris["Species"] == "virginica"]["Petal.Length"].median(), color='green', linestyle='solid', linewidth=2, alpha = 0.5)
plt.legend(prop={'size': 16}, title = 'IRIS PETAL LENGTH')
plt.title('Density Plot of 3 Iris Species Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Density')
```
